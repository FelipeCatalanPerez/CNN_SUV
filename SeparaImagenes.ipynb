# Separar data y crear carpetas

```python
import glob
import shutil
import os

import pandas as pd
import numpy as np
SEED=43   
```

```python
df = pd.read_csv('Clase_SUV.csv')
df = df.iloc[:,[1,2]]
df
```

```python
X, y = df.iloc[:, 0:1], df.iloc[:, [1]]

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=SEED)

## Para SUV
# TEST 20%
y_test_SUV = y_test[y_test["Dummy"]==1]
X_test1_SUV = X_test[y_test["Dummy"]==1]

# TRAIN 80%
y_train1_SUV = y_train[y_train["Dummy"]==1]
X_train1_SUV = X_train[y_train["Dummy"]==1]

## Para NO SUV
# TEST 20%
y_test1_NOSUV = y_test[y_test["Dummy"]==0]
X_test1_NOSUV = X_test[y_test["Dummy"]==0]

# TRAIN 80%
y_train1_NOSUV = y_train[y_train["Dummy"]==0]
X_train1_NOSUV = X_train[y_train["Dummy"]==0]
```

```python
X_test1_SUV
```

```python
#1
files0 = X_test1_SUV['Imagen']

# Hay que crear una carpeta vacia que se llame dest_folder

for f in files0:
    shutil.copy(f, 'SUV_test')
```

```python
#2
files0 = X_train1_SUV['Imagen']

# Hay que crear una carpeta vacia que se llame dest_folder

for f in files0:
    shutil.copy(f, 'SUV_train')
```

```python
#3
files0 = X_test1_NOSUV['Imagen']

# Hay que crear una carpeta vacia que se llame dest_folder

for f in files0:
    shutil.copy(f, 'NO_SUV_test')
```

```python
#4
files0 = X_train1_NOSUV['Imagen']

# Hay que crear una carpeta vacia que se llame dest_folder

for f in files0:
    shutil.copy(f, 'NO_SUV_train')
```
